# Analysis of ANOVA and Pairwise T-Test Results

---

## 1. Overview of Analysis Objectives

The purpose of this analysis is to:
- Test whether there are significant differences in **average speeds** across various transportation modes.
- Identify which specific pairs of modes exhibit significant differences through **pairwise T-tests**.


---

## 2. ANOVA Results Summary

The **ANOVA** test evaluates if the means of average speeds differ across all transportation modes. The results are:

- **F-Statistic**: **669.662**
- **P-Value**: **0.000e+00**

**Conclusion**:  
Since the p-value is significantly less than 0.05 (**p < 0.05**), we reject the null hypothesis.  
This confirms that **there are statistically significant differences in the average speeds across the transportation modes**.

---

## 3. Pairwise T-Test Results Summary

The pairwise **T-tests** identify specific pairs of modes that exhibit statistically significant differences in average speeds. Below is the summary of pairwise comparisons:

| **Mode 1**                        | **Mode 2**                        | **T-Statistic** | **P-Value**   |
|----------------------------------|----------------------------------|----------------|--------------|
| Car / Motorbike                  | Walking + Public Transport       | 98.343         | 1.27e-47     |
| Car / Motorbike                  | Public Transport                 | 93.204         | 2.31e-197    |
| Car / Motorbike                  | Walking                          | 79.303         | 1.22e-146    |
| Car / Motorbike                  | Bicycle                          | -0.328         | 7.44e-01     |
| Walking + Public Transport       | Public Transport                 | 44.816         | 9.18e-10     |
| Walking + Public Transport       | Walking                          | 38.102         | 6.55e-26     |
| Walking + Public Transport       | Bicycle                          | -0.742         | 6.14e-01     |
| Public Transport                 | Walking                          | 49.421         | 1.08e-13     |
| Public Transport                 | Bicycle                          | 48.204         | 1.25e-92     |
| Walking                          | Bicycle                          | 0.531          | 6.14e-01     |
| Walking + Public Transport       | Walking + Public Transport + Bicycle | -1.654         | 1.53e-01     |
| Public Transport + Car / Motorbike | Walking + Public Transport     | 45.786         | 1.26e-14     |

**Key Findings**:
- **Car / Motorbike** has significant differences compared to **Walking**, **Walking + Public Transport**, and **Public Transport**.
- **Bicycle** shows **no significant differences** when compared to **Walking** or **Car / Motorbike** (P-Value > 0.05).
- **Walking + Public Transport** has no significant difference with combinations involving **Bicycle**.

---

## 4. Key Observations from Pairwise Comparisons

### Car / Motorbike vs Bicycle
- **P-Value**: 0.744
- **Conclusion**: No significant difference in average speeds.  
**Insight**: The speeds for bicycles and cars/motorbikes are not drastically different in this dataset.

### Walking vs Bicycle
- **P-Value**: 0.614
- **Conclusion**: No significant difference in average speeds.  
**Insight**: Walking and cycling speeds appear to be relatively similar.

### Walking vs Car / Motorbike
- **P-Value**: ~0 (close to zero)
- **Conclusion**: Significant difference in average speeds.  
**Insight**: Cars/motorbikes move significantly faster than walking.

---

## 5. General Insights

- **Modes with Significant Speed Differences**:
   - Car/Motorbike consistently exhibits significantly higher average speeds compared to walking and public transport.
   - Public Transport also has significantly different speeds when compared to walking or mixed modes.

- **Modes without Significant Speed Differences**:
   - Bicycle and walking speeds do not differ significantly.
   - Combined modes (e.g., "Walking + Public Transport") often do not show significant improvements over single modes like walking.

---

## 6. Conclusion

The results confirm that **significant differences** exist in average speeds across most transportation modes, particularly between high-speed modes (e.g., Car/Motorbike) and lower-speed modes (e.g., Walking, Bicycling).

The pairwise T-test heatmap effectively highlights these differences, offering clear insights for decision-makers in urban transportation planning.
