# Analysis of Transport Modes, Speed, and Distance


## **1. Average Speed by Transport Mode**

**Insights:**
- Car / Motorbike achieves the highest average speed (~13 km/h), significantly outperforming other modes.
- Modes like Public Transport and Walking show considerably lower average speeds (~3–5 km/h).
- Bicycle speeds remain consistent (~6 km/h), but speeds drop significantly in mixed modes, such as Walking + Bicycle or Walking + Public Transport + Bicycle.

**Interpretation:**
- Car / Motorbike dominates in speed, highlighting its efficiency in urban or suburban travel settings.
- Public Transport speeds are lower due to frequent stops, delays, and reduced flow during peak hours.
- Mixed modes that involve walking suffer from overall lower average speeds.

---

## **2. Distribution of Distance by Mode of Transport**

**Insights:**
- Car / Motorbike trips cover the largest distances, with a median of ~8–10 km and several trips exceeding 25 km.
- Walking distances are understandably short, typically under 2 km, with minimal variability.
- Bicycle trips are moderately consistent, covering distances between 2–5 km, with some long outliers reaching up to ~15 km.
- Public Transport and mixed modes demonstrate shorter distances but include occasional longer trips.

**Interpretation:**
- Car / Motorbike remains the preferred mode for long-distance travel, with its higher speed and reliability.
- Walking trips are constrained by physical limits, while Bicycle serves as a medium-distance option.
- The variability in Public Transport distances highlights its adaptability for both short and long-distance travel.

---

## **3. Frequency of Trips by Mode of Transport**

**Insights:**
- Car / Motorbike and Bicycle are the most frequent modes of travel, reflecting their popularity for regular commutes.
- Public Transport is also widely used, suggesting accessibility and affordability for travelers.
- Mixed modes, such as Walking + Public Transport are less frequent, indicating limited adoption or fewer scenarios requiring intermodal transport.

**Interpretation:**
- Urban commuters likely prefer Car / Motorbike due to flexibility, efficiency, and speed.
- Bicycle serves as an eco-friendly and economical choice for short to medium distances.

---

## **4. Frequency of Trips by Purpose**

**Insights:**
- The most frequent trip purpose is "Returning to my home", which aligns with commuters traveling back after work or errands.
- Work / Business Appointment follows, emphasizing trips related to employment.
- Less frequent trip purposes include "Health," "Education," and "Dropping off / Picking up someone", indicating occasional rather than routine travel.

**Interpretation:**
- Trip purpose distributions reveal that urban travel patterns are predominantly home-to-work or home-to-errand-based.
- Lower-frequency purposes like Health and Education suggest specialized but critical travel behaviors.

---

## **5. Distribution of Speed by Mode of Transport**

**Insights:**
- Car / Motorbike has the highest median speeds (~12–15 km/h) but shows variability due to traffic conditions and road types.
- Bicycle speeds are clustered tightly around 6 km/h, highlighting its consistent performance.
- Public Transport speeds have a lower median and significant variability due to stops, transfers, and delays.
- Mixed modes involving walking and public transport demonstrate very low speeds (~2–4 km/h), consistent with walking interludes.

**Interpretation:**
- Car / Motorbike ensures efficient travel in terms of speed, particularly in areas with minimal congestion.
- Public Transport requires improvements to reduce variability and increase reliability.
- Mixed modes suffer from low speeds, pointing to inefficiencies caused by walking transitions.
